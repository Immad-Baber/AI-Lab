```python
import random

environment = {
    "A": random.choice(["Clean", "Dirty"]),
    "B": random.choice(["Clean", "Dirty"])
}

internal_state = {
    "A": "Unknown",
    "B": "Unknown"
}


location = "A"

print("Initial Environment:", environment)
print()

for step in range(1, 6):
    print("Step",step)

    percept = environment[location]

    internal_state[location] = percept

    if internal_state[location] == "Dirty":
        action = "Clean"
        environment[location] = "Clean"
    else:
        action = "Move"
        location = "B" if location == "A" else "A"

    print("Agent Location:", location)
    print("Percept:", percept)
    print("Action:", action)
    print("Internal State:", internal_state)
    print("Environment:", environment)
    print()
```

# Question no 3


```python
routes = {
    "Route A": {"distance": 10, "time": 15, "risk": "Low"},
    "Route B": {"distance": 8, "time": 20, "risk": "Medium"},
    "Route C": {"distance": 12, "time": 10, "risk": "High"}
}

risk_values = {
    "Low": 1,
    "Medium": 2,
    "High": 3
}

utility_scores = {}

for route_name, route in routes.items():
    distance = route["distance"]
    time = route["time"]
    risk = risk_values[route["risk"]]

    utility_score = (distance + time + risk)
    utility_scores[route_name] = utility_score

best_route = min(utility_scores, key=utility_scores.get)
print("____________________________________________")
print("Utility Scores:")
for route, score in utility_scores.items():
    print(route, ":", score)

print("\nSelected Route:", best_route)
```
