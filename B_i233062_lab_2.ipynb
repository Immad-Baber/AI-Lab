# Question no 7 

```python

import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")

Q1 = tips[['total_bill', 'tip']].quantile(0.25)
Q3 = tips[['total_bill', 'tip']].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print("____________________________________________________")
print("Lower Bounds\n", lower)

print("\nUpper Bounds\n", upper)
print("____________________________________________________")

outliers = ((tips[['total_bill', 'tip']] < lower) |(tips[['total_bill', 'tip']] > upper))

outlier_count = outliers.sum()
print("____________________________________________________")

print("\nNumber of Outliers per Column:\n", outlier_count)

tips.boxplot(column=['total_bill'])
plt.title("Box Plot: Total Bill")
plt.show()

tips.boxplot(column=['tip'])
plt.title("Box Plot: Tip")
plt.show()
```

# Question no 8

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

print("Dataset size before outlier removal:", tips.shape)

Q1 = tips[['total_bill', 'tip']].quantile(0.25)
Q3 = tips[['total_bill', 'tip']].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

tips_cleaned = tips[
    (tips['total_bill'] >= lower_bound['total_bill']) &(tips['total_bill'] <= upper_bound['total_bill']) &
    (tips['tip'] >= lower_bound['tip']) & (tips['tip'] <= upper_bound['tip'])
]

print("Dataset size after outlier removal:", tips_cleaned.shape)


tips.boxplot(column=['total_bill'])
plt.title("Box Plot Before Outlier Removal")
plt.show()

tips.boxplot(column=['tip'])
plt.title("Box Plot Before Outlier Removal")
plt.show()

tips_cleaned.boxplot(column=['total_bill'])
plt.title("Box Plot After Outlier Removal")
plt.show()

tips_cleaned.boxplot(column=['tip'])
plt.title("Box Plot After Outlier Removal")
plt.show()
```
